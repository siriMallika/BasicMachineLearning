```python
import pandas as pd
import numpy as np
import sklearn

print(pd.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
from sklearn import datasets
iris  = datasets.load_iris()
print('Data: ')
print(iris['data'].shape)
print('Target: ')
print(iris['target'].shape)
```

```python
print('Sample of Data')
print(iris['data'][0:5])
print('Sample of Target')
print(iris['target'][0:5])
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(iris['data'],iris['target'])
from sklearn.model_selection import cross_val_score
cross_val_score(model, iris['data'], iris['target'], cv=10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris['data'],
iris['target'],test_size = 0.3)
model_2 = DecisionTreeClassifier()
model_2.fit(x_train,y_train)

from sklearn.metrics import accuracy_score

y_pred = model_2.predict(x_train)
print('Accuracy of Training set: ')
print(accuracy_score(y_train,y_pred))

y_pred = model_2.predict(x_test)
print('Accuracy of Test set: ')
print(accuracy_score(y_test,y_pred))
```

```python
dataset = [['Milk','Orion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Milk','Apple', 'Kidney Beans' ,'Eggs'],
           ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
           ['Corn','Onion','Onion','Kidney Beans','Ice cream' ,'Eggs']
          ]
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df,min_support = 0.6)
```

```python
apriori(df,min_support = 0.6,use_colnames=True)
```
